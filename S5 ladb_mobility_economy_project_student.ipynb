Resumen ejecutivo

Contexto y Objetivo
Este análisis busca determinar la relación entre la movilidad urbana (retrasos por congestión y tiempos de viaje) y la productividad económica (PIB per cápita) para entender si la riqueza de una ciudad influye en su tráfico. Cobertura de Datos
El estudio procesó un volumen masivo de información inicial con más de 1,000,000 de registros en el dataset de tráfico y 30 registros en el económico. El análisis final se centró específicamente en el año 2024, consolidando datos de diversas ciudades globales y países (como Argentina, Brasil, Colombia, Emiratos Árabes y España) para garantizar la relevancia temporal de los hallazgos.

Metodología
